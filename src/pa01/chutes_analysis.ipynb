```python
from snakes_and_ladders import multi_game_experiment

import matplotlib.pyplot as plt


players = [1, 2, 4, 8]

for player in players:
    
    num_moves = multi_game_experiment(100, player, 79)
    
    plt.hist(num_moves, histtype='step', bins=int(max(num_moves) / 1.5))
    plt.title(f'Distribution of game durations with {player} player(s)')
    plt.xlabel('Duration of game (number of moves to win)')
    plt.ylabel('Frequency count')
    plt.show()
```

### Our analysis
From the histograms we see that when there are more players in the game the mean duration of the game is shorter, and the variance decreases.
